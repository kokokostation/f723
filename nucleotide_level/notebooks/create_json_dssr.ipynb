```python
import os
from tqdm import tqdm
```

```python
CIF_DIR = '/home/mikhail/bioinformatics/data/NR_3.0/cif'
JSON_DIR = '/home/mikhail/bioinformatics/data/NR_3.0/json'
```

```python
for fname in tqdm(os.listdir(CIF_DIR)):
    pdb_id, _ = os.path.splitext(fname)
    
    input_path = os.path.join(CIF_DIR, fname)
    output_path = os.path.join(JSON_DIR, '{}.json'.format(pdb_id))
    
    ! x3dna-dssr --input=$input_path --format=mmcif --idstr=long --u-turn --more \
    --non-pair --po4 --output=$output_path --json
```
